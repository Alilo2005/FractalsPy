```python
import numpy as np
import matplotlib.pyplot as plt

# Julia function
def julia(c, xmin, xmax, ymin, ymax, width, height, max_iter):
    x_vals = np.linspace(xmin, xmax, width)
    y_vals = np.linspace(ymin, ymax, height)
    X, Y = np.meshgrid(x_vals, y_vals)
    Z = X + 1j * Y
    M = np.zeros(Z.shape, dtype=int)

    for i in range(max_iter):
        mask = np.abs(Z) < 2
        M[mask] = i
        Z[mask] = Z[mask] ** 2 + c

    return M

# Generate and plot
julia_set = julia(complex(-0.8, 0.156), -1.5, 1.5, -1.5, 1.5, 1000, 1000, 100)

fig, ax = plt.subplots(figsize=(6, 6), facecolor="black")
ax.set_facecolor("black")
ax.imshow(julia_set, cmap="plasma", extent=(-1.5, 1.5, -1.5, 1.5))
ax.set_xticks([])
ax.set_yticks([])
ax.axis("off")
ax.set_title("Julia Set", color="white", fontsize=14)
plt.show()
```
