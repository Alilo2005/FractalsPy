```python
import numpy as np
import matplotlib.pyplot as plt

# Mandelbrot function
def mandelbrot(xmin, xmax, ymin, ymax, width, height, max_iter):
    x_vals = np.linspace(xmin, xmax, width)
    y_vals = np.linspace(ymin, ymax, height)
    X, Y = np.meshgrid(x_vals, y_vals)
    C = X + 1j * Y
    Z = np.zeros_like(C, dtype=np.complex64)
    M = np.zeros(C.shape, dtype=int)

    for i in range(max_iter):
        mask = np.abs(Z) < 2
        M[mask] = i
        Z[mask] = Z[mask] ** 2 + C[mask]

    return M

# Generate and plot
mandelbrot_set = mandelbrot(-2, 1, -1.5, 1.5, 1000, 1000, 100)

fig, ax = plt.subplots(figsize=(6, 6), facecolor="black")
ax.set_facecolor("black")
ax.imshow(mandelbrot_set, cmap="inferno", extent=(-2, 1, -1.5, 1.5))
ax.set_xticks([])
ax.set_yticks([])
ax.axis("off")
ax.set_title("Mandelbrot Set", color="white", fontsize=14)
plt.show()
```
