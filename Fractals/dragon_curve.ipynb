```python
import matplotlib.pyplot as plt

# Recursive Dragon Curve function
def dragon_curve(order, p1, p2, points):
    if order == 0:
        points.append(p1)
        points.append(p2)
        return
    mid = ((p1[0] + p2[0]) / 2 + (p2[1] - p1[1]) / 2,
           (p1[1] + p2[1]) / 2 - (p2[0] - p1[0]) / 2)
    dragon_curve(order - 1, p1, mid, points)
    dragon_curve(order - 1, p2, mid, points)

# Generate and plot
points = []
dragon_curve(12, (0, 0), (1, 0), points)
points = list(dict.fromkeys(points))  # Remove duplicates

fig, ax = plt.subplots(figsize=(6, 6), facecolor="black")
ax.set_facecolor("black")
ax.plot(*zip(*points), color="red", lw=0.5)
ax.set_xticks([])
ax.set_yticks([])
ax.axis("off")
ax.set_title("Dragon Curve", color="white", fontsize=14)
plt.show()
```
