```python
import numpy as np
import matplotlib.pyplot as plt

# Ensure inline plotting
%matplotlib inline  

# Transformation coefficients for Barnsley Fern
coefficients = [
    (0.0, 0.0, 0.0, 0.16, 0.0, 0.0, 0.01),
    (0.85, 0.04, -0.04, 0.85, 0.0, 1.6, 0.85),
    (0.2, -0.26, 0.23, 0.22, 0.0, 1.6, 0.07),
    (-0.15, 0.28, 0.26, 0.24, 0.0, 0.44, 0.07)
]

# Function to generate Barnsley Fern points
def generate_fern(n):
    x, y = 0, 0
    points_x, points_y = [], []
    
    for _ in range(n):
        r = np.random.random()
        total = 0
        for a, b, c, d, e, f, p in coefficients:
            total += p
            if r <= total:
                x, y = a*x + b*y + e, c*x + d*y + f
                points_x.append(x)
                points_y.append(y)
                break
    
    return points_x, points_y

# Generate points
num_points = 50000  
x_vals, y_vals = generate_fern(num_points)

# Create the figure with a black background
fig, ax = plt.subplots(figsize=(6, 10), dpi=300, facecolor='black')  # Higher resolution
ax.set_facecolor('black')  # Set plot background color to black
ax.scatter(x_vals, y_vals, s=0.1, color="lime", alpha=0.8)  # Use bright green points

# Remove axis labels and ticks
ax.set_xticks([])
ax.set_yticks([])
ax.axis("off")

# Unicode math-style title
title_text = (
    "Barnsley Fern\n"
    "xₙ₊₁ = a ⋅ xₙ + b ⋅ yₙ + e\n"
    "yₙ₊₁ = c ⋅ xₙ + d ⋅ yₙ + f"
)
ax.set_title(title_text, color="white", fontsize=14, pad=20)

# Force display
plt.draw()
plt.pause(0.01)
plt.show(block=True)
```
