```python
import numpy as np
import matplotlib.pyplot as plt

# Points of the main triangle
vertices = np.array([[0, 0], [1, 0], [0.5, np.sqrt(3) / 2]])

# Generate Sierpiski Triangle points
def sierpinski(n):
    points = []
    p = np.random.rand(2)  # Start with a random point
    for _ in range(n):
        v = vertices[np.random.randint(0, 3)]  # Random vertex
        p = (p + v) / 2  # Midpoint rule
        points.append(p)
    return np.array(points)

# Generate and plot
n_points = 50000
points = sierpinski(n_points)

fig, ax = plt.subplots(figsize=(6, 6), facecolor="black")
ax.set_facecolor("black")
ax.scatter(points[:, 0], points[:, 1], s=0.1, color="cyan")
ax.set_xticks([])
ax.set_yticks([])
ax.axis("off")
ax.set_title("Sierpiski Triangle", color="white", fontsize=14)
plt.show()
```
